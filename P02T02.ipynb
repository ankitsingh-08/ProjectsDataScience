```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv(r"C:\Users\ankit\OneDrive\Documents\TrainData.csv")
test_data = pd.read_csv(r"C:\Users\ankit\OneDrive\Documents\TestData.csv")
```

```python
train_data.describe()
```

```python
train_data.isnull().sum()
```

```python
train_data = train_data.dropna(axis = 0)
```

```python
test_data.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
model = LinearRegression()
```

```python
x_train = train_data['x']
y_train = train_data.drop('x', axis = 1)
```

```python
x_test = test_data['x']
y_test = test_data.drop('x', axis = 1)
```

```python
plt.scatter(x_train, y_train, s=5)
plt.xlabel("X...")
plt.ylabel("Y>>>")
```

```python
chart = sns.regplot(x = x_train, y = y_train, scatter_kws = {'s':30, 'facecolor':'white'}, line_kws = {'color':'red'})
chart.set_title("Regression Of Y on X", fontdict = {'size':20})
plt.show()
```

```python
model.fit(y_train, x_train)
```

```python
m = model.intercept_
c = model.coef_
m,c
```

```python
y_hat = model.predict(y_test)
```

```python
y_test.shape
```

```python
index = [i for i in range(1,301)]
plt.plot(index,y_test, color="blue", linewidth = 2, linestyle = '-')
plt.plot(index,y_hat, color="yellow", linewidth = 1, linestyle = '-')
plt.title('Given BY Predicted')
plt.xlabel('Index')
plt.ylabel('Y>>>')
plt.xticks(np.arange(0,240, step = 36))
plt.yticks(np.arange(0,96, step = 24))
plt.grid()
```

```python
from sklearn import metrics
```

```python
print("MSE = ", metrics.mean_squared_error(y_hat, y_test))
print("r^2 = ", metrics.r2_score(y_hat, y_test))
```

COMPARING THE VALUES OF MSE AND r^2,WE CAN SAY OUR MODEL IS 
99.98% ACCURATE ON OUR TEST DATA!
